__Exercise 2.__ Modify the query above to return the gene name in yeast, the gene names in the other three species, the cai value and the sequence for all yeat genes with a cai value larger that 0.8 and that have orthologs in the other three species.

```python
pd.read_sql('select title, mikatae, paradoxus, bayanus, cai, seq FROM orthologs LEFT JOIN cai ON orthologs.cerevisiae = cai.title LEFT JOIN sequences on orthologs.cerevisiae = sequences.gene_name WHERE cai > 0.8' , con)
```

__Exercise 3.__ Perform the same query as above, but this time include the chromosome number of the gene in the output.

```python
pd.read_sql('select title, chrom, mikatae, paradoxus, bayanus, cai, seq FROM orthologs LEFT JOIN cai ON orthologs.cerevisiae = cai.title LEFT JOIN sequences on orthologs.cerevisiae = sequences.gene_name left join gff on cai.title = gff.acc  WHERE cai > 0.8' , con)
```

__Exercise 4.__ Now generate a table with the number of genes per chromosome that have CAI larger than 0.5 and have orthologs in the other three yeast species.

```python
pd.read_sql('select chrom, count(chrom) FROM orthologs LEFT JOIN cai ON orthologs.cerevisiae = cai.title LEFT JOIN sequences on orthologs.cerevisiae = sequences.gene_name left join gff on cai.title = gff.acc  WHERE cai > 0.5 group by chrom' , con)
```
